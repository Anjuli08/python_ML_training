```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('jerry.jpg')

img2 = cv2.imread('3.png')
img = img[:, 0:1080,:]
```

```python
img.shape, img2.shape
```

```python
def showing(im):
    plt.imshow(im[:,:,::-1])
    plt.show()

def imshow(im):
    cv2.imshow('Preview', im)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
```

```python
#imshow(img)
```

```python
showing(img)
showing(img2)
```

```python
subtracted = cv2.subtract(img, img2)
showing(subtracted)
```

```python
import numpy as np
```

```python
img2_1 = cv2.multiply(img2, 2*np.ones(img2.shape).astype('uint8'))
showing(img2_1)
```

```python
img2_2 = cv2.convertScaleAbs(img2, alpha = 1, beta = 80)
showing(img2_2)
```

```python
img2_3 = cv2.resize(img2, (400,400), interpolation=cv2.INTER_CUBIC)
img2_3.shape
```

```python
img2.shape
```

```python
showing(img2_3)
```

```python
img2_4 = cv2.resize(img2, None, fx=2, fy=0.5, interpolation=cv2.INTER_CUBIC)
img2_4.shape
```

```python
showing(img2_4)
```
